```python
%%writefile playlist.h
#ifndef PLAYLIST_H
#define PLAYLIST_H

#include <iostream>
#include <string>
using namespace std;

// Forward declaration of SongPlaylist class
template <typename T>
class SongPlaylist;

// Forward declaration of the function template
template <typename T>
void shuffleSongs(SongPlaylist<T>&);

template <typename T>
class Playlist {
    protected:
        struct Node {
            T songName, singer, songLength, genre;
            Node* prev;
            Node* next;
            Node(T name, T singer, T length, T genre)
                : songName(name), singer(singer), songLength(length), genre(genre), prev(nullptr), next(nullptr) {}
        };
        Node* head;
        Node* tail;
        Node* current;
        bool isSorted;
    public:
        Playlist() : head(nullptr), tail(nullptr), current(nullptr), isSorted(true) {}
        ~Playlist() {
            while (head) {
                Node* temp = head;
                head = head->next;
                delete temp;
            }
        }
        virtual void sortSongs() = 0;
        friend void shuffleSongs<>(SongPlaylist<T>&);
};

#endif
```

```python
%%writefile songplaylist.h
#ifndef SONGPLAYLIST_H
#define SONGPLAYLIST_H

#include <string>
#include "playlist.h"
using namespace std;

template <typename T>
class SongPlaylist : public Playlist<T> {
public:
    // Constructor
    SongPlaylist();
    // Destructor
    ~SongPlaylist();
    // Function prototypes
    void initializeList(const T&);
    void addSong(const T&, const T&, const T&, const T&);
    T tolower(const T&);
    void nextSong();
    void prevSong();
    void writeToFile(const T&) const;
    void destructList();
    void sortSongs() override;
    void displayCurrentSong() const;
    void displayAllSongs() const;
    typename Playlist<T>::Node* searchSong(const T&);
    void searchByGenre(const T& genre) const;
    void editCurrentSong(const T&, const T&, const T&, const T&);
    void deleteNode(typename Playlist<T>::Node*);
    typename Playlist<T>::Node* getCurrentPointer() const;
};

#endif
```

```python
%%writefile songplaylist.cpp
#include <iostream>
#include <fstream>
#include <string>
#include "songplaylist.h"
using namespace std;

const int MAX_GENRES = 999;

// Doubly Linked List class to manage the playlist
template <typename T>
SongPlaylist<T>::SongPlaylist() : Playlist<T>() {}

template <typename T>
SongPlaylist<T>::~SongPlaylist() {}

template <typename T>
void SongPlaylist<T> :: initializeList(const T& filename) {
    ifstream file(filename);
    if (!file) {
        throw filename;
        return;
    }
    int i = 0;
    string header, line;
    getline(file, header);
    while (getline(file, line)) {
        size_t pos1 = line.find('\t');
        size_t pos2 = line.find('\t', pos1 + 1);
        size_t pos3 = line.find('\t', pos2 + 1);
        if (pos1 != string::npos && pos2 != string::npos && pos3 != string::npos) {
            string songName = line.substr(0, pos1);
            string singer = line.substr(pos1 + 1, pos2 - pos1 - 1);
            string songLength = line.substr(pos2 + 1, pos3 - pos2 - 1);
            string genre = line.substr(pos3 + 1);
            addSong(songName, singer, songLength, genre);
            i++;
        }
    }
    file.close();
    cout << i << " songs loaded and sorted from file: " << filename << "\n";
    this->current = this->head;
    if(i == 0){
        cout << "No songs found.\n";
    }
}

// Lowercasing so it can compare and sort when adding songs
template <typename T>
T SongPlaylist<T>::tolower(const T& songName) {
    T lowerStr = songName;
    for (int i = 0; i < lowerStr.length(); ++i) {
        if(lowerStr[i] >= 'A' && lowerStr[i] <= 'Z'){
            lowerStr[i] += 32; // convert to lowercase
        }
    }
    return lowerStr;
}

// Add a song to the playlist
template <typename T>
void SongPlaylist<T>::addSong(const T& songName, const T& singer, const T& songLength, const T& genre) {
    typename Playlist<T>::Node* newNode = new typename Playlist<T>::Node(songName, singer, songLength, genre);
    typename Playlist<T>::Node* temp = this->head;
    typename Playlist<T>::Node* prev = nullptr;

    if (!this->head) {
        this->head = this->tail = newNode;
    } else {
        while(temp){
            if(temp->songName == songName && temp->singer == singer){
                cout << "Song already exists in the playlist.\n";
                return;
            }
            temp = temp->next;
        }
        // Ordered insertion, insert by song name A-Z
        temp = this->head;
        while (temp && (tolower(temp->songName) < tolower(songName))) {
            prev = temp;
            temp = temp->next;
        }
        if (!temp || !this->isSorted) {
            // Append to the end of the list
            // if the song name is the last
            this->tail->next = newNode;
            newNode->prev = this->tail;
            this->tail = newNode;
        }
        else if (temp == this->head) {
            // Insert at the beginning of the list
            newNode->next = this->head;
            this->head->prev = newNode;
            this->head = newNode;
        }
        else {
            // Insert in the middle of the list
            newNode->next = temp;
            newNode->prev = prev;
            prev->next = newNode;
            temp->prev = newNode;
        }
    }
    cout << songName << " " << singer << " " << genre << " (" << songLength << " mins) added to the playlist.\n";
}

template <typename T>
void SongPlaylist<T>::nextSong() {
    if (!this->current || !this->current->next) {
        cout << "No next song.\n";
        return;
    }
    this->current = this->current->next;
    cout << "Playing " << this->current->songName << " " << this->current->singer << " " << this->current->genre << " (" << this->current->songLength << " mins)\n";
}

// Move to the previous song
template <typename T>
void SongPlaylist<T>::prevSong() {
    if (!this->current || !this->current->prev) {
        cout << "No previous song.\n";
        return;
    }
    this->current = this->current->prev;
    cout << "Playing " << this->current->songName << " " << this->current->singer << " " << this->current->genre << " (" << this->current->songLength << " mins)\n";
}

template <typename T>
void SongPlaylist<T>::writeToFile(const T& filename) const {
    ofstream file(filename);
    if (!file) {
        throw filename;
        return;
    }

    file << "Song Name\tAuthor\tSong Length\tGenre\n";

    typename Playlist<T>::Node* temp = this->head;
    while (temp) {
        file << temp->songName << '\t' << temp->singer << '\t' << temp->songLength << '\t' << temp->genre << '\n';
        temp = temp->next;
    }

    file.close();
    cout << "Playlist written to file: " << filename << "\n";
}

template <typename T>
void SongPlaylist<T>::sortSongs(){
    typename Playlist<T>::Node* temp = this->head;
    while (temp) {
        typename Playlist<T>::Node* min = temp;
        typename Playlist<T>::Node* temp2 = temp->next;
        while (temp2) {
            if (tolower(temp2->songName) < tolower(min->songName)) {
                min = temp2;
            }
            temp2 = temp2->next;
        }
        if (min != temp) {
            T tempName = temp->songName;
            T tempSinger = temp->singer;
            T tempLength = temp->songLength;
            T tempGenre = temp->genre;
            temp->songName = min->songName;
            temp->singer = min->singer;
            temp->songLength = min->songLength;
            temp->genre = min->genre;
            min->songName = tempName;
            min->singer = tempSinger;
            min->songLength = tempLength;
            min->genre = tempGenre;
        }
        temp = temp->next;
    }
}

template <typename T>
void SongPlaylist<T>::destructList() {
    try {
        writeToFile("song.txt");
    } catch (string filename) {
        cout << "ERROR: Cannot write to file " << filename << endl;
    }
    typename Playlist<T>::Node* temp = this->head;
    while (temp) {
        typename Playlist<T>::Node* next = temp->next;
        delete temp;
        temp = next;
    }
    this->head = this->tail = this->current = nullptr;
    cout << "Playlist destructed.\n";
}

// Display the current song
template <typename T>
void SongPlaylist<T>::displayCurrentSong() const {
    if (!this->current) {
        cout << "\nNo song is currently playing.\n";
        return;
    }
    cout << "\n======================= Currently Playing =======================\n";
    cout << this->current->songName << " " << this->current->singer << " " << this->current->genre << " (" << this->current->songLength << " mins)\n";
    cout << "=================================================================\n";
}

template <typename T>
void SongPlaylist<T>::displayAllSongs() const {
    typename Playlist<T>::Node* temp = this->head;
    int count = 1;
    cout << "\n======================= Playlist =======================\n";
    while (temp) {
        cout << count << ". " << temp->songName << " " << temp->singer << " " << temp->genre << " (" << temp->songLength << " mins)\n";
        temp = temp->next;
        count++;
    }
    if(count == 1){
        cout << "No songs in the playlist.\n";
    }
    cout << "=======================================================\n";
}

template <typename T>
typename Playlist<T>::Node* SongPlaylist<T>::searchSong(const T& songName) {
    typename Playlist<T>::Node* temp = this->head;
    while (temp) {
        if (temp->songName == songName) {
            cout << songName << " found in the playlist.\n";
            this->current = temp; // after song found, assign it as currently playing
            return temp;
        }
        temp = temp->next;
    }
    cout << songName << " not found in the playlist.\n";
    return nullptr;
}

template <typename T>
void SongPlaylist<T>::searchByGenre(const T& genre) const {
    typename Playlist<T>::Node* temp = this->head;
    bool found = false;
    cout << "\n Your personal " << genre << " playlist\n=================================================================\n";
    while (temp) {
        if (temp->genre == genre) {
            cout << temp->songName << " " << temp->singer << " " << temp->genre << " (" << temp->songLength << " mins)\n";
            found = true;
        }
        temp = temp->next;
    }
    if (!found) {
        cout << "No songs found in the genre " << genre << ".\n";
    }
    cout << "=================================================================\n";
}

// Edit the content of the current node
template <typename T>
void SongPlaylist<T>::editCurrentSong(const T& newSongName, const T& newSongLength, const T& newSinger, const T& newGenre) {
    typename Playlist<T>::Node* temp = this->head;
    while(temp){
        // added temp != current to ignore/skip the current same song name, user might wants to edit the current song's details
        if(temp->songName == newSongName && temp->singer == newSinger && temp != this->current){
            cout << "Song already exists in the playlist.\n";
            return;
        }
        temp = temp->next;
    }
    if (this->current) {
        this->current->songName = newSongName;
        this->current->singer = newSinger;
        this->current->songLength = newSongLength;
        this->current->genre = newGenre;
        cout << "\nCurrent song edited to: " << newSongName << " " << newSinger << " " << newGenre << " (" << newSongLength << " mins)\n";
    }
    else {
        cout << "No current song to edit.\n";
    }
}

template <typename T>
void SongPlaylist<T>::deleteNode(typename Playlist<T>::Node* node) {
    if (!node) return;

    if (node == this->head) {
        this->head = node->next;
        if (this->head) this->head->prev = nullptr;
    }
    else {
        node->prev->next = node->next;
    }

    if (node == this->tail) {
        this->tail = node->prev;
        if (this->tail) this->tail->next = nullptr;
    }
    else {
        node->next->prev = node->prev;
    }

    if (node == this->current) {
        this->current = node->next ? node->next : node->prev;
    }

    delete node;

    cout << "This song is deleted from the playlist.\n";
}

template <typename T>
typename Playlist<T>::Node* SongPlaylist<T>::getCurrentPointer() const{
    if (!this->current) {
        string noCurrent = "No current song! ";
        throw noCurrent;
    }
    return this->current;
}

template class SongPlaylist<string>; // instantiating the class
```

```python
%%writefile main.cpp
#include <iostream>
#include "playlist.h"
#include "songplaylist.h"
using namespace std;

const int MAX_STREAM_SIZE = 32767;
void displayMenu();
bool exitConfirmation();
bool validateInput(int,int,int);
bool validateInputPureDigit(const string&);
bool validateInputString(const string&);
bool validateSongLength(const string&);

int main() {
    SongPlaylist<string> myPlaylist;
    bool running = true;

    try {
        myPlaylist.initializeList("song.txt");
    } catch (string filename) {
        cout << "ERROR: Cannot open file " << filename << endl;
        running = false;
    }

    while (running) {

        myPlaylist.displayCurrentSong();

        int choice;
        do {
            displayMenu();
            cin >> choice;
        } while (validateInput(choice,1,10));

        cin.ignore(MAX_STREAM_SIZE, '\n'); // Clear input buffer
        string songName, genre,editSongName, editSongLength, editSinger, editGenre, newSongName, newSongLength, newSinger, newGenre;
        switch (choice) {
        case 1:
            myPlaylist.nextSong();
            break;
        case 2:
            myPlaylist.prevSong();
            break;
        case 3:
            do{
                cout << "Enter new song name: ";
                getline(cin, editSongName);
            }while(validateInputString(editSongName));
            do{
                cout << "Enter new song length: ";
                getline(cin, editSongLength);
            }while(validateSongLength(editSongLength));
            do{
                cout << "Enter new singer: ";
                getline(cin, editSinger);
            }while(validateInputString(editSinger));
            do{
                cout << "Enter new genre: ";
                getline(cin, editGenre);
            }while(validateInputString(editGenre) || validateInputPureDigit(editGenre));
            myPlaylist.editCurrentSong(editSongName, editSongLength, editSinger, editGenre);
            break;

        case 4:
            try{
                myPlaylist.deleteNode(myPlaylist.searchSong(myPlaylist.getCurrentPointer()->songName));
            }catch(string noCurrent){
                cout << noCurrent << "Nothing to delete.\n";
            }
            break;

        case 5:
            do{
                cout << "Enter song name to search: ";
                getline(cin, songName);
            }while(validateInputString(songName));
            myPlaylist.searchSong(songName);
            break;

        case 6:
            do{
                cout << "What genre of music are you in the mood for right now?\n";
                cin >> genre;
            }while(validateInputString(genre) || validateInputPureDigit(genre));
            myPlaylist.searchByGenre(genre);
            break;

        case 7:
            do{
                cout << "Enter new song name: ";
                getline(cin, newSongName);
            }while(validateInputString(newSongName));
            do{
                cout << "Enter new song length: ";
                getline(cin, newSongLength);
            }while(validateSongLength(newSongLength));
            do{
                cout << "Enter new singer: ";
                getline(cin, newSinger);
            }while(validateInputString(newSinger));
            do{
                cout << "Enter new genre: ";
                getline(cin, newGenre);
            }while(validateInputString(newGenre) || validateInputPureDigit(newGenre));
            myPlaylist.addSong(newSongName, newSinger, newSongLength, newGenre);
            break;

        case 8:
            myPlaylist.displayAllSongs();
            break;

        case 9:
            try {
                shuffleSongs<string>(myPlaylist);
                cout << "Playlist shuffled.\n";
            } catch (string noShuffle) {
                cout << noShuffle;
            }
            myPlaylist.displayAllSongs();
            break;

        case 10:
            if (exitConfirmation()) {
                running = false;
                myPlaylist.sortSongs(); // sort songs before rewriting to the file
                myPlaylist.destructList();
                cout << "Program ended.";
            }
            break;
        default:
            break;
        }
    }

    return 0;
}


void displayMenu() {
    cout << "\n======== Song Playlist Menu ========\n"
         << "  1. Next Song\n"
         << "  2. Previous Song\n"
         << "  3. Edit Current Song\n"
         << "  4. Delete Current Song\n"
         << "  5. Search for a Song\n"
         << "  6. Song Recommendation By Genre\n"
         << "  7. Add Songs\n"
         << "  8. Display All Songs\n"
         << "  9. Shuffle Songs\n"
         << "  10. Exit\n"
         << "====================================\n"
         << "Enter your choice: ";
}

bool exitConfirmation() {
    string isExit;
    cout << endl;
    cout << "Press Y/y to exit program: ";
    cin >> isExit;
    cin.ignore(MAX_STREAM_SIZE, '\n');
    if (isExit == "Y" || isExit == "y") {
        return true;
    }
    return false;
}

bool validateInput(int userInput, int min, int max) {
    if (!(cin.fail()) && userInput >= min && userInput <= max) {
        return false;
    } else {
        cout << endl;
        cout << "ERROR: Input must be from " << min << " to " << max << "." << endl;
        cout << endl;
        // since the input might be char for int type, so we need to clear the error flag
        cin.clear();
        // and clear any input buffer
        cin.ignore(MAX_STREAM_SIZE, '\n');
        return true;
    }
}

bool validateInputPureDigit(const string& userInput) {
    for (int i = 0; i < userInput.length(); i++) {
        if (!isdigit(userInput[i])) {
            // Not digit, return false to break while loop (correct input)
            return false;
        }
    }
    // All digits
    cout << "ERROR: Input must not be all digits.\n";
    return true;
}

bool validateInputString(const string& userInput) {
    // Check if the string is empty
    if (userInput.empty()) {
        cout << "ERROR: Input cannot be empty.\n";
        return true;
    }
    // Check if first char for each word is capital letter or digit
    for(int i = 0; i < userInput.length(); i++) {
        // first char for each word
        if((i == 0 || userInput[i-1] == ' ') && !isdigit(userInput[i])){
            // not capital letter?
            if (!isupper(userInput[i])) {
                cout << "ERROR: First character for each word must be a capital letter.\n";
                return true;
            }
        }
        if(i == userInput.length() - 1 && userInput[i] == ' '){
            cout << "ERROR: There must be no space at the end.\n";
            return true;
        }
    }
    return false;
}

bool validateSongLength(const string& songLength) {
    if (songLength.empty()) {
        cout << "ERROR: Input cannot be empty.\n";
        return true;
    }
    size_t colonPos = songLength.find(':');
    if (colonPos == string::npos || colonPos == 0 || colonPos == songLength.length() - 1) {
        // If there is no colon, or the colon is at the start or end of the string, the format is invalid
        cout << "ERROR: Invalid format. Please enter the song length in the format mm:ss.\n";
        return true;
    }

    for (int i = 0; i < songLength.length(); ++i) {
        if (i != colonPos && !isdigit(songLength[i])) {
            // If any character other than the colon is not a digit, the format is invalid
            cout << "ERROR: Invalid format. Please enter the song length with DIGITS in the format mm:ss.\n";
            return true;
        }
    }

    string minutes, seconds;
    // Extract minutes and seconds
    for(int i = 0; i < songLength.length(); i++){
        if(i < colonPos){
            minutes += songLength[i];
        }else if(i > colonPos){
            seconds += songLength[i];
        }
    }

    // Check if minutes and seconds are within valid range
    if (minutes.length() != 2 || seconds.length() != 2 || minutes < "00" || minutes > "59" || seconds < "00" || seconds > "59") {
        cout << "ERROR: Invalid time or format. Minutes and seconds must be between 00 and 59 in the format mm:ss.\n";
        return true;
    }

    return false;
}

template <typename T>
void shuffleSongs(SongPlaylist<T>& playlist) {
    // get the size of the linked list
    int size = 0;
    for (typename Playlist<T>::Node* node = playlist.head; node != nullptr; node = node->next) {
        size++;
    }

    if(size != 0){
        // shuffle the linked list
        for (typename Playlist<T>::Node* node = playlist.head; node != nullptr; node = node->next) {
            // Get a random index from 0 to size - 1
            int j = rand() % size;

            // Traverse the linked list to find the song at index j
            typename Playlist<T>::Node* song_j = playlist.head;
            for (int k = 0; k < j; ++k) {
                song_j = song_j->next;
            }

            // Swap the songs
            typename Playlist<T>::Node temp = {node->songName, node->singer, node->songLength, node->genre};
            node->songName = song_j->songName;
            node->singer = song_j->singer;
            node->songLength = song_j->songLength;
            node->genre = song_j->genre;
            song_j->songName = temp.songName;
            song_j->singer = temp.singer;
            song_j->songLength = temp.songLength;
            song_j->genre = temp.genre;
        }
    }else{
        string noShuffle = "No songs to shuffle.\n";
        throw noShuffle;
    }

    // Set the current song to the head of the list
    playlist.current = playlist.head;

    // Set the isSorted flag to false
    playlist.isSorted = false;
}
```
